```python
import pandas as pd
import re
from tqdm import tqdm_notebook
```

```python
raw_data = pd.read_csv('train', sep='\n', header=None)
```

```python
#detach id


train['id']= [raw_data[0][i].partition('\t')[0] for i in range(len(raw_data))]
train['else']= [raw_data[0][i].partition('\t')[2] for i in range(len(raw_data))]
train['label'] = [[] for _ in range(len(train))]
```

```python
#detach labels


train['label'] = [[train['else'][j].partition('\t')[0]] for j in range(len(train)) ]
train['else'] = [train['else'][j].partition('\t')[2] for j in range(len(train))]   
```

```python
#check for multiple labels if they exist and detach it


for i in range(4):
    mask = train['else'].str.contains('__label__')
    print(sum(mask))
    for j in tqdm(range(len(train))):
        if mask[j]:
            train['label'][j].append(train['else'][j].partition('\t')[0])
            train['else'][j] = train['else'][j].partition('\t')[2]       
```

```python
#get rid of __label__ in labels


for i in tqdm(range(len(train))):
    train['label'][i] = [string[9:] for string in train['label'][i]]
```

```python
import pandas as pd
from tqdm import tqdm_notebook
```

```python
train = pd.read_csv('splitted_train.csv')
train['label'] = [eval(i) for i in train['label']]
train['else'] = [eval(i) for i in train['else']]
```

```python
#normalising words


import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
#normalising words
for i in tqdm_notebook(range(len(train))):
    for j in range(len(train['else'][i])):
        train['else'][i][j] = morph.parse(train['else'][i][j])[0].normal_form
```
